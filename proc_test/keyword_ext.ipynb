```python
import sys
sys.path.insert(0, '../')
import auth_prop
import pymysql
from datetime import datetime

mysql_auth = auth_prop.mysql

# MySQL Connection 연결
con = pymysql.connect(host=mysql_auth['host'], user=mysql_auth['user'], password=mysql_auth['passwd']
                      , db=mysql_auth['db'], charset=mysql_auth['charset'], port=mysql_auth['port'])

# Connection 으로부터 Cursor 생성
cur = con.cursor()

# SQL문 실행 및 Fetch
sql = """SELECT NEWSITEMID , NEWS_CNTS
FROM ABKL_NEWS_MST anm
WHERE DATE_FORMAT(REGIST_DT, '%%Y-%%m-%%d') = DATE_FORMAT(%s, '%%Y-%%m-%%d')
    AND NEWS_CNTS != '%%{[contents][0]}'
ORDER BY REGIST_DT
LIMIT 10
"""
cur.execute(sql, "2023-01-09")
# cur.execute(sql, str(datetime.today())[:10])

# 데이타 Fetch
rows = cur.fetchall()
print(rows)

# STEP 5: DB 연결 종료
# con.close()
```

```python
from python_api.keyword_module import keyword_ext

# SQL문 실행 및 Fetch
sql = "INSERT IGNORE INTO ABKL_NEWS_KEYWORD_RESULT(NEWSITEMID, KEYWORD_SEQ, KEYWORD) VALUES(%s,%s,%s)"

for row in rows:
    # 기사 본문 전처리
    text = row[1].replace("<![CDATA[","").replace("]]>","")
    # 키워드 추출
    res = keyword_ext(text)
    for idx,keyword in enumerate(res):
        cur.execute(sql, (row[0], str(idx+1), keyword))
    con.commit()
cur.close()
con.close()
```
